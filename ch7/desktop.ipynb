```python
for w in p.getAllWindows():
    print(w)
```

```python
for w in p.getWindowsWithTitle("제목 없음"):
    print(w)
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]

# 만약에 창이 활성화가 안되어있다면 활성화를 시켜주세욤
print(w)
if not w.isActive:
    w.activate()
```

```python
# 만약에 창이 최대화가 아니라면 최대화를 시켜주세욤
if not w.isMaximized:
    w.maximize()

# 만약에 창이 최소화가 아니라면 최소화를 시켜주세욤
if not w.isMinimized:
    w.minimize()
```

```python
# 창 원래대로 돌려주세욤
w.restore()
```

```python
# 창 종료
w.close()
```

```python
# pip install pyperclip
import pyperclip

# 메모장 가져오기
w = p.getWindowsWithTitle("제목 없음")[0]
# 메모장 활성화
w.activate()

# 클립보드 복사
pyperclip.copy("파이썬")

# 클립보드 내용 붙여넣기
p.hotkey("ctrl","v")
```

```python
def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
word_write("amvlksmdklfma")
word_write("졸려워욤")
```
